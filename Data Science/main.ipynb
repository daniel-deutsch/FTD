```python
import requests
import pandas as pd
```

```python
url = "https://api.stocktwits.com/api/2/streams/symbol/AAPL.json"
r = requests.get(url)
data = r.json()
```

## Level 1

```python
for msg in data['messages']:
    print(msg['body'])
```

## Level 2

```python
for msg in data['messages']:
    print(msg['created_at'], msg['user']['username'], msg['body'])
```

## Level 3

```python
for msg in data['messages']:
    if msg['entities']['sentiment']:
        print(msg['entities']['sentiment']['basic'])
    else:
        print(msg['entities']['sentiment'])
```

## Level 4

```python
avg = 0
for msg in data['messages']:
    if msg['entities']['sentiment']:
        if msg['entities']['sentiment']['basic'] == 'Bullish':
            avg += 1
        elif msg['entities']['sentiment']['basic'] == 'Bearish':
            avg -= 1

avg /= len(data['messages'])
print(f"{avg=}")
```

## Level 5

```python
df = pd.DataFrame()
for msg in data['messages']:
    df = df.append({
        'date': msg['created_at'],
        'username': msg['user']['username'],
        'content': msg['body'],
        'sentiment': msg['entities']['sentiment']['basic'] if msg['entities']['sentiment'] else None
    }, ignore_index=True)

df
```
